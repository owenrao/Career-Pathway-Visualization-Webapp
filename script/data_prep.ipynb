## Explore and Clean Data

```python
## open merged data
raw_data = '/content/drive/My Drive/DSCI551/Final_Project/Data/Raw Data'
oews_df = pd.read_csv(raw_data + '/2019-2020_oews.csv')
```

```python
#remove unncesessary features
oews_df = oews_df.drop(['Unnamed: 0','area','i_group','pct_total','prim_state'], axis=1)

#we also potentially drop: area_type, naics_title, own_code, hourly, year
```

```python
oews_df.dtypes
```

```python
#manually filtering numerical columns into list
col_str = 'tot_emp	emp_prse	jobs_1000	loc_quotient	pct_total	h_mean	a_mean	mean_prse	h_pct10	h_pct25	h_median	h_pct75	h_pct90	a_pct10	a_pct25	a_median	a_pct75	a_pct90'
num_features = [column for column in oews_df.columns if column in col_str]
num_features
```

```python
#filling missing values (*) with NaNs and # with max values (100.0 for hourly and 208000 for annual)
for column in num_features:
  oews_df[column] = oews_df[column].replace('*',np.nan)
  oews_df[column] = oews_df[column].replace('**',np.nan)
  if column.startswith('h'):
    oews_df[column] = oews_df[column].replace('#',100.0)
  elif column.startswith('a'):
    oews_df[column] = oews_df[column].replace('#', 208000)
  oews_df[column] = oews_df[column].astype(float) # cast all numerics to floats
```
